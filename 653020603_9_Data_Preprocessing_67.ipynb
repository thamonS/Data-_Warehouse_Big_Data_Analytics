#2

* (10 points) จากข้อมูลที่ให้มา เตรียมข้อมูลให้พร้อมโดยการจัดการกับค่าว่างในข้อมูล ให้อธิบายวิธีที่ท่านเลือกใช้ พร้อมยกเหตุผลประกอบ

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
titanic['fare'].fillna(titanic['fare'].median(), inplace=True)
titanic['cabin'].fillna('Unknown', inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)
titanic['boat'].fillna('Unknown', inplace=True)
titanic['body'].fillna('Unknown', inplace=True)
```

```python
titanic
```

##Answer

>**age** ใช้ค่า mean เพื่อหาค่าเฉลี่ย

>**fare** ใช้ค่า median มาแทนค่าว่างเพื่อหาราคาที่อยู่กึ่งกลางของราคาทั้งหมด จะไม่ใช้ mean แทนที่ค่าว่างเพื่อไม่ให้มันกลายเป็น outlier

>**embarked** ใช้ค่า mode เพราะคอลัมน์นี้เป็นข้อมูลที่แบ่งหมวดหมู่ของท่าเรือ

>**cabin, boat, body** ใช้การแทนที่ค่าว่างเป็น Unknown เพราะคอลัมน์ดังกล่าวเป็น string  

#3

* แปลงข้อมูลเชิงคุณภาพให้เป็นตัวเลข \
    i. (10 points) ใช้ one-hot encoding ในการแปลงตัวแปร Embarked \
    ii. (10 points) แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1 \
    iii. (10 points) สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard) \
    iv. (10 points) สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

##i. ใช้ one-hot encoding ในการแปลงตัวแปร Embarked

```python
embarked_encoded = pd.get_dummies(titanic['embarked'], prefix='embarked',dtype=int)
titanic = pd.concat([titanic, embarked_encoded], axis=1)
titanic
```

##ii. แปลงตัวแปร Sex โดยให้เพศชายเป็น 0 และเพศหญิงเป็น 1

```python
sex_trf = titanic['sex'].map({'male':'0', 'female':'1'})
titanic['sex'] = sex_trf
titanic
```

##iii. สร้างตัวแปร FamilySize ซึ่งเป็นผลรวมของตัวแปร SibSp (number of siblings/spouses aboard) และ Parch (number of parents/children aboard)

```python
titanic['FamilySize'] = titanic['sibsp']+titanic['parch']
titanic
```

##iv. สร้างตัวแปร IsAlone ซึ่งมีค่าเป็น 1 ถ้า FamilySize เป็น 0, และมีค่าเป็นศูนย์เมื่อเป็นอื่น

```python
titanic['IsAlone'] = [1 if i == 0 else 0 for i in titanic['FamilySize']]
titanic
```

* ทำการ Normalize ข้อมูล \
    i. (10 points) สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)\
    ii. (10 points) สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1 \
    iii. (10 points) ตัวแปร Fare และ Age ใช้ Min-Max Scaling

```python
import pandas as pd
import numpy as np
pd.cut(np.array([1, 7, 5, 4, 6, 3]), 3)
```

##i.ตัวแปร Fare และ Age ใช้ Min-Max Scaling

```python
!pip install scikit-learn
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
titanic[['fare','age']] = scaler.fit_transform(titanic[['fare', 'age']])
titanic
```

##ii. สร้างตัวแปรใหม่ Fare_Per_Person ด้วยการนำตัวแปร Fare มาหารด้วย FamilySize + 1


```python
titanic['Fare_Per_Person'] = titanic['fare']/(titanic['FamilySize'] + 1)
titanic
```

##iii. สร้างตัวแปร AgeGroup ที่ระบุกลุ่มอายุของตัวอย่าง โดยแบ่งเป็นกลุ่มดังนี้ [0, 12), [12, 18), [18, 50) และ [50, 100)

```python
titanic['AgeGroup'] = pd.cut(titanic['age'], bins=[0,12,18,50,100], labels=['[0, 12)','[12, 18)','[18, 50)','[50, 100)'])
titanic
```

* เมื่อเตรียมข้อมูลพร้อมแล้ว ให้หาความสัมพันธ์ของตัวแปรต่อไปนี้จากข้อมูล ซึ่งอาจนำเสนอด้วยสถิติและแผนภูมิหรือกราฟก็ได้: \
    i. (10 points) survived และ pclass \
    ii. (10 points) survived และ sex \
    iii. (10 points) survived และ age \
    iv. (10 points) จำนวนสมาชิกครอบครัวที่รอดชีวิต และ Fare

##i. survived และ pclass
ดูจำนวนผู้โดยสารที่เสียชีวิตและรอดชีวิตโดยแบ่งกลุ่มตามชั้นที่นั่ง

```python
pivot_survived_pclass = titanic.pivot_table(values='pclass', index='survived', aggfunc='mean')
pivot_survived_pclass
pd.crosstab(titanic['survived'], titanic['pclass']).plot(kind='bar')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

>จากกราฟจะเห็นว่าผู้โดยสารที่ไม่รอดชีวิตส่วนใหญ่จะเป็นผู้โดยสารที่นั่งชั้นที่ 3 และ ผู้โดยสารที่รอดชีวิตส่วนใหญ่จะเป็นผู้โดยสารที่นั่งชั้นที่ 1 และ 2

##ii. survived และ sex

```python
titanic.pivot_table(values='survived', index='sex', aggfunc='mean')
```

```python
pd.crosstab(titanic['survived'], titanic['sex']).plot(kind='bar')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(['male','female'],loc='upper right')
plt.show()
```

>จากผล pivot ด้านบนจะเห็นว่าเพศหญิงมีโอกาสรอดชีวิตมากกว่าเพศชาย
